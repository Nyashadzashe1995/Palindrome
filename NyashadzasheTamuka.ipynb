### ~What is the average NoPLHIV of the two estimates used for “Xhariep”?

```python
# lets try to extract the data where estimate column is survey or Fay-Heriott and store in survey_df and Fay_Heriott_df variable
survey_df = df[df['Estimate']=='Survey']
Fay_Heriott_df = df[df['Estimate']=='Fay-Heriott']
```

```python
# Now, we extract the data where District is Xhariep according to survey and store the data into variable s.
s = survey_df[survey_df['District']=='Xhariep']
```

```python
# lets try to check the average of  NoPLHIV by using mean function and store in avg_s variable
avg_s = s['NoPLHIV'].mean()
print("average NoPLHIV according to estimate(Survey) :",avg_s)
```

```python
# Now, we extract the data where District is Xhariep according to Fay_Heriott and store the data into variable f.
f = Fay_Heriott_df[Fay_Heriott_df['District']=='Xhariep']
```

```python
# lets try to check the average of NoPLHIV by using mean function and store in avg_f variable
avg_f = f['NoPLHIV'].mean()
print("average NoPLHIV according to estimate(Fay_Heriott) :",avg_f)
```

### ~Add a column and populate it with the number of people not living with HIV for each row.

```python
'''
Now, we create a new column name People and store all the data of people not living with HIV.
We create a list comprehension and iterate all the estimate column value if our estimate values is Survey
then we store the dash and if our estimate value is another then we store the value 'Not Living'.
'''
df['People'] = ['-' if i=='Survey' else 'Not Living' for i in df['Estimate']]
```

```python
'''Now, we can see that our new column is add in our original dataframe and also we can see that where estimate value is not 
surevy we set the value Not living in People column'''
df.head(20)
```

### ~What is the total NoPLHIV in all the cities (districts with the word “city” or “metro” in the name)?

```python
'''
Now, we extracting the data where city and metro name is in district column.
we create a comprehension list and iterate all the district and then we convert all the 
district in lower letter and split the words in list by using split function in python and then check the city or 
metro in list that we have split by using split function if exist then store into city or metro variable.
'''
city = [c for c in df['District'] if 'city' in c.lower().split()]
metro = [m for m in df['District'] if 'metro' in m.lower().split()]
```

```python
# lets try to check the district where city name is exist
city
```

```python
# lets try to check the total NoPLHIV with the word city
print("total NoPLHIV with the word city in a district :",len(city))
```

```python
# lets try to check the district where metro name is exist
metro
```

```python
# lets try to check the total NoPLHIV with the word metro
print("total NoPLHIV with the word metro in a district :",len(metro))
```
